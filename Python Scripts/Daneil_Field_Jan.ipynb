```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
weather = pd.read_csv("Daneil Field Monthly.csv", index_col="DATE")
```

```python
rolling_mean = january_weather['prcp'].rolling(window=5, min_periods=1).mean()
z = np.polyfit(mdates.date2num(january_weather.index), january_weather['prcp'], 4)
p = np.poly1d(z)

# Plotting
plt.figure(figsize=(15, 7))  # Set a suitable figure size
plt.plot(january_weather.index, january_weather['prcp'], 'o', color='blue', linestyle='--', markersize=5, alpha=0.5, label='Daily Precipitation')  # Scatter plot for daily data
plt.plot(january_weather.index, rolling_mean, color='red', linestyle='-', linewidth=2, label='5-Year Rolling Mean')  # Rolling mean line
plt.plot(january_weather.index, p(mdates.date2num(january_weather.index)), 'b-', label='Trend Line')  # Trend line

# Optional: Highlight certain statistical metrics or ranges
plt.axhline(y=january_weather['prcp'].mean(), color='blue', linestyle='--', label='Overall Mean')

# Fill between the rolling mean and some boundary or another statistical line if needed
plt.fill_between(january_weather.index, rolling_mean, january_weather['prcp'].mean(), color='green', alpha=0.1)

# Styling and labels
plt.title('Daneil Field Station January Precipitation Analysis (1999-2024)')
plt.xlabel('Year')
plt.ylabel('Precipitation (inches)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
rolling_mean = january_weather['prcp'].rolling(window=5, min_periods=1).mean()

# Calculate the linear fit coefficients
linear_fit_coeffs = np.polyfit(mdates.date2num(january_weather.index), january_weather['prcp'], 1)
linear_fit = np.poly1d(linear_fit_coeffs)

# Plotting
plt.figure(figsize=(15, 7))  # Set a suitable figure size
plt.plot(january_weather.index, january_weather['prcp'], 'o', color='blue', linestyle='-', markersize=5, alpha=0.5, label='Daily Precipitation')  # Scatter plot for daily data
plt.plot(january_weather.index, rolling_mean, color='green', linestyle='-', linewidth=2, label='5-Year Rolling Mean')  # Rolling mean line


# Add linear fit line
plt.plot(january_weather.index, linear_fit(mdates.date2num(january_weather.index)), color='red', linestyle='-', linewidth=2, label='Trend Line')

# Optional: Highlight certain statistical metrics or ranges
plt.axhline(y=january_weather['prcp'].mean(), color='blue', linestyle='--', label='Overall Mean')

# Fill between the rolling mean and some boundary or another statistical line if needed
plt.fill_between(january_weather.index, rolling_mean, january_weather['prcp'].mean(), color='green', alpha=0.1)

# Styling and labels
plt.title('Daneil Field January Precipitation Analysis (1999-2024)')
plt.xlabel('Year')
plt.ylabel('Precipitation (inches)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
weather_q1 = weather[weather.index.month.isin([1, 2, 3])]

# Group the data by year and sum the precipitation for bar plot
annual_q1_precip = weather_q1.resample('Y').sum()

# Calculate linear fit for the annual sums
years = mdates.date2num(annual_q1_precip.index)  # Convert dates to numerical format
slope, intercept = np.polyfit(years, annual_q1_precip['prcp'], 1)
trend_line = slope * years + intercept

# Calculate 5-year rolling mean
rolling_mean = annual_q1_precip['prcp'].rolling(window=5, min_periods=1).mean().dropna()

# Overall mean
overall_mean = annual_q1_precip['prcp'].mean()

# Ensure indices match for x and y values
annual_years = annual_q1_precip.index.year  # Only years where data exists
rolling_years = rolling_mean.index.year

# Plotting
plt.figure(figsize=(14, 7))  # Set a suitable figure size

# Plotting bars for combined Q1 precipitation for each year
plt.bar(annual_years, annual_q1_precip['prcp'], color='skyblue', width=0.6, label='Q1 Total Precipitation')

# Plot individual month data points
for month in [1, 2, 3]:
    monthly_data = weather[weather.index.month == month]
    plt.scatter(monthly_data.resample('Y').sum().index.year, monthly_data.resample('Y').sum()['prcp'], label=f'Month {month} Data')

# Plotting the trend line (ensure the years match)
plt.plot(annual_years, trend_line, color='red', linestyle='-', linewidth=2, label='Linear Trend Line')

# Plotting the 5-year rolling mean (match years)
plt.plot(rolling_years, rolling_mean, color='green', linestyle='-', linewidth=2, label='5-Year Rolling Mean')

# Plotting the overall mean
plt.axhline(y=overall_mean, color='blue', linestyle='--', linewidth=2, label='Overall Mean')

# Add labels and title
plt.title('Annual Q1 Precipitation Analysis')
plt.xlabel('Year')
plt.ylabel('Total Precipitation (inches)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
seasonal_weather = weather[(weather.index.month == 12) | (weather.index.month <= 2)& (weather.index.year>1998)].copy()
```

```python
seasonal_weather
```

```python
# Adjust year for December to be part of the next year's winter
seasonal_weather['adjusted_year'] = seasonal_weather.index.year
seasonal_weather.loc[seasonal_weather.index.month == 12, 'adjusted_year'] += 1
```

```python
winter_precip = seasonal_weather.groupby('adjusted_year')['prcp'].sum()
```

```python
winter_precip
```

```python
# Setup for plotting
years = np.array(winter_precip.index)  # Years as x-values
precip_values = winter_precip.values

# Calculate the linear trend line
slope, intercept = np.polyfit(years, precip_values, 1)
trend_line = slope * years + intercept

# Calculate 5-year rolling mean
#rolling_mean = winter_precip.rolling(window=5, min_periods=1).mean()

# Calculate overall mean
overall_mean = winter_precip.mean()

# Plotting
plt.figure(figsize=(13, 6))
plt.bar(years, precip_values, color='skyblue', label='Total Winter Precipitation')

# Plotting trend line and rolling mean
plt.plot(years, trend_line, 'r-', label='Trend Line')
#plt.plot(years, rolling_mean, 'g-', label='5-Year Rolling Mean')

# Plot overall mean
plt.axhline(y=overall_mean, color='blue', linestyle='--', linewidth=2, label='Overall Mean')

tick_interval = 2  # Adjust this value as needed for optimal spacing
plt.xticks([year for year in years if year % tick_interval == 0])

plt.title('Daniel Field Winter Season Precipitation Analysis')
plt.xlabel('Winter Season Ending Year')
plt.ylabel('Precipitation (inches)')
plt.legend()
plt.grid(True)
plt.show()
```
